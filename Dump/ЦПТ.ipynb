```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
poisson_rv = sts.poisson(3)
sample=poisson_rv.rvs(1000)
```

```python
print sample
```

```python
plt.hist(sample,normed=True)
```

```python
x=np.linspace(1,10,20)
pmf=poisson_rv.pmf(x)
plt.plot(x,pmf)
```

```python
poisson_rv = sts.poisson(5)
poisson_rv.rvs(10)
```

```python
x = np.linspace(0,30,31)
for l in [1, 5, 10, 15]:
    rv = sts.poisson(l)
    cdf = rv.cdf(x)
    plt.step(x, cdf, label="$\lambda=%s$" % l)
plt.legend()
plt.title("CDF (poisson)")

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
norm_rv=sts.norm(3)
# на том же графике построим теоретическую плотность распределения:
x = np.linspace(0,30,31)
pmf = norm_rv.pdf(x)
plt.plot(x, pmf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
